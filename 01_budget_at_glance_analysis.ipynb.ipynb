```python
clean_exp = clean_exp[[1,2,3,4,5]]

clean_exp.columns = [
    "category",
    "actual_2024_25",
    "budget_2025_26",
    "revised_2025_26",
    "budget_2026_27"
]

clean_exp
```

```python
clean_exp = clean_exp[
    ~clean_exp["category"].isin([
        "A. Centre's Expenditure",
        "B. Transfers",
        "Grants/Loans/Transfers",
        "Grand Total"
    ])
]

clean_exp
```

```python
clean_exp["category"] = clean_exp["category"].replace({
    "II Central Sector": "Central Sector Schemes",
    "Schemes/Projects": "Central Sector Schemes",
    "IV Centrally Sponsored": "Centrally Sponsored Schemes",
    "Schemes": "Centrally Sponsored Schemes",
    "III Other Central Sector": "Other Central Sector Expenditure",
    "VI Other": "Other Expenditure"
})

clean_exp
```

```python
clean_exp["category"] = clean_exp["category"].replace({
    "II Central Sector": "Central Sector Schemes",
    "Schemes/Projects": "Central Sector Schemes",
    "IV Centrally Sponsored": "Centrally Sponsored Schemes",
    "Schemes": "Centrally Sponsored Schemes",
    "III Other Central Sector": "Other Central Sector Expenditure",
    "VI Other": "Other Expenditure"
})

clean_exp
```

```python
capability_map = {
    "Establishment Expenditure": "Governance",
    "Central Sector Schemes": "Economic Opportunity",
    "Other Central Sector Expenditure": "Economic Opportunity",
    "Interest Payments": "Non-Capability",
    "Centrally Sponsored Schemes": "Social Protection",
    "Finance Commission Grants": "Social Protection",
    "Other Expenditure": "Other"
}

clean_exp["capability_category"] = clean_exp["category"].map(capability_map)

clean_exp
```

```python
capability_summary = clean_exp.groupby("capability_category")[[
    "actual_2024_25",
    "budget_2025_26",
    "revised_2025_26",
    "budget_2026_27"
]].sum()

capability_summary
```

```python
capability_summary = capability_summary.astype(float)
```

```python
total_budget_2026 = capability_summary["budget_2026_27"].sum()

capability_summary["share_2026_budget"] = (
    capability_summary["budget_2026_27"] / total_budget_2026
) * 100

capability_summary
```

```python
import matplotlib.pyplot as plt

capability_summary["share_2026_budget"].plot(
    kind="bar",
    figsize=(8,5)
)

plt.title("Share of Union Budget 2026 by Capability Category")
plt.ylabel("Percentage of Budget")
plt.xlabel("Capability Category")

plt.show()
```

```python
capability_summary[[
    "actual_2024_25",
    "budget_2025_26",
    "revised_2025_26",
    "budget_2026_27"
]].T.plot(figsize=(10,6))

plt.title("Capability Spending Trends in Union Budget")
plt.ylabel("Expenditure (₹ Crore)")
plt.xlabel("Budget Year")

plt.show()
```

## Policy Interpretation and Capability Analysis

## 1. Introduction

The "Budget at a Glance" document provides a summarized overview of the Union Budget of India. It presents key aggregates of government receipts and expenditures along with important fiscal indicators such as fiscal deficit, revenue deficit, effective revenue deficit, and primary deficit.

The document also illustrates the sources of government receipts and the composition of government expenditure through tables and graphical representations.

According to the Budget at a Glance 2026–27 document, the total expenditure of the Government of India in Budget Estimates 2026–27 is ₹53,47,315 crore. Out of this, the total capital expenditure is estimated at ₹12,21,821 crore and the effective capital expenditure is estimated at ₹17,14,523 crore.

## 2. Structure of Government Expenditure

The Budget at a Glance document classifies government expenditure into multiple broad components.

These include:

• Establishment Expenditure  
• Central Sector Schemes / Projects  
• Other Central Sector Expenditure  
• Centrally Sponsored Schemes  
• Finance Commission Grants  
• Other Grants, Loans and Transfers  

Together, these components constitute the total expenditure of the Government of India.

## 3. Data Preparation and Classification

For the purpose of analysis, expenditure data from the Budget at a Glance document was organized into broader functional categories.

The dataset includes the following budget stages:

• Actuals 2024–25  
• Budget Estimates 2025–26  
• Revised Estimates 2025–26  
• Budget Estimates 2026–27  

Expenditure components were grouped into the following capability categories:

• Economic Opportunity  
• Social Protection  
• Other  

This grouping allows the analysis to examine the relative distribution of government expenditure across broader policy categories.

## 4. Aggregation of Expenditure by Capability Category

Using Python and the Pandas library, expenditure data was grouped by capability category.

The groupby() function was used to aggregate expenditure values across the four budget stages.

This step produces total expenditure for each category, allowing comparison between categories and across budget years.

## 5. Calculation of Budget Share

After aggregating expenditure values, the share of each category in the total Union Budget 2026–27 was calculated.

The formula used for this calculation is:

Share of Category = (Category Expenditure / Total Expenditure) × 100

This metric helps evaluate the relative importance of different expenditure categories within the overall budget.

## 6. Visualization of Budget Structure

Two visualizations were created using the Matplotlib library.

The first visualization presents the share of each capability category in the Union Budget 2026–27 using a bar chart.

The second visualization shows the trend in expenditure across different budget years. This allows comparison between Actuals, Budget Estimates, and Revised Estimates.

## 7. Observations from the Analysis

The analysis shows that a significant share of expenditure falls under the category of Economic Opportunity.

A smaller share of expenditure is classified under Social Protection, while the remaining expenditure is categorized as Other.

This distribution highlights the relative allocation of government spending across different functional categories within the Union Budget.

## 8. Conclusion

This analysis provides a structured overview of the Union Budget by grouping expenditure into broader functional categories and examining their relative share in total expenditure.

By combining data aggregation and visualization techniques, the analysis highlights the structure and distribution of government expenditure across different categories.
